```python
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt  # has basic diagrams, non-interative
import seaborn as sns            # has graphical tools
```

```python
os.chdir('C:/Personals/GWU/202103_Programming for Analytics II_DNSC_6303_10/Assignment_3/')
```

```python
df1 = pd.read_csv("Pandemic.csv")
df1['ds']=pd.to_datetime(df1['ds'], dayfirst = True) # convert the object data in date datatype
df1= df1.sort_values('ds')
df2 = df1[df1['location'] == 'Andorra' ]
fig, ax = plt.subplots(figsize =(10,10))
ax.set_title('Condition of covid-19 cases of covid-19 in Andorra')
ax.plot(df2[df2['data_type'] == 'deaths_new' ]['ds'],df2[df2['data_type'] == 'deaths_new' ]['value'],label='deaths_new', color = 'blue')
ax.plot(df2[df2['data_type'] == 'cases_new' ]['ds'],df2[df2['data_type'] == 'cases_new' ]['value'],label='cases_new', color = 'orange')
ax.plot(df2[df2['data_type'] == 'recovered_new' ]['ds'],df2[df2['data_type'] == 'recovered_new' ]['value'],label='recovered_new', color = 'green')
plt.xlabel("Months")
plt.ylabel("Covid cases('Deaths,cases and Recovered')")
plt.legend()
plt.show()
```
